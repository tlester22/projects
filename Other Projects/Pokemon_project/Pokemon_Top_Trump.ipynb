```python
# first, we need to import our data from API and input the question
import random
import requests
pokemon_id = input("What Pokemon you will choose from 1 to 151?")
url = "https://pokeapi.co/api/v2/pokemon/{}", pokemon_id

# we need to define our random choice. actually, we can change 151 to bigger number if you want
def random_choice():
  return random.randint(1,151)

# define chosen_id as a value for our statement
chosen_id = int(pokemon_id)

# adding a condition to have statement if you choose our number
if 1 < chosen_id < 151:
    print ("Nice choice!")
else:
    print ("Please choose from 1 to 151")
```

```python
def pokemon_player_selection():
    '''Will run within battle script below, allows players to choose their own pokemon'''
    
    player_choice_1 = int(input('Choose you first Pokemon from 1 to 151.'))
    player_choice_2 = int(input('Choose your second Pokemon from 1 to 151?'))
    
    if 1 < player_choice_1 < 151 and 1 < player_choice_1 < 151:
        print ("Nice team!")
    else:
        print ("Please choose from 1 to 151")
    
    
    url_player_1 = "https://pokeapi.co/api/v2/pokemon/{}".format(player_choice_1)
    url_player_2 = "https://pokeapi.co/api/v2/pokemon/{}".format(player_choice_2)
    
    response_player_1 = requests.get(url_player_1)
    response_player_2  = requests.get(url_player_2)
    
    player_pokemon_1 = response_player_1.json()
    player_pokemon_2 = response_player_2.json()
    
    player_stats_1 = { 
        'name': player_pokemon_1['name'], 
        'id': player_pokemon_1['id'], 
        'height': player_pokemon_1['height'], 
        'weight': player_pokemon_1['weight'], 
        } 
    player_stats_2 = { 
        'name': player_pokemon_2['name'], 
        'id': player_pokemon_2['id'], 
        'height': player_pokemon_2['height'], 
        'weight': player_pokemon_2['weight'], 
        } 
    
    return [player_stats_1, player_stats_2]

pokemon_player_selection()
```

```python
def pokemon_opponent_selection():
    '''Will run within battle script below, chooses random pokemon for CPU opponent'''
    
    opponent_choice_1 = random.randint(1,151)
    opponent_choice_2 = random.randint(1,151)
    
    url_opponent_1 = "https://pokeapi.co/api/v2/pokemon/{}".format(opponent_choice_1)
    url_opponent_2 = "https://pokeapi.co/api/v2/pokemon/{}".format(opponent_choice_2)
    
    response_opponent_1 = requests.get(url_opponent_1)
    response_opponent_2 = requests.get(url_opponent_2)
    
    opponent_pokemon_1 = response_opponent_1.json()
    opponent_pokemon_2 = response_opponent_2.json()
    
    opponent_stats_1 = { 
        'name': opponent_pokemon_1['name'], 
        'id': opponent_pokemon_1['id'], 
        'height': opponent_pokemon_1['height'], 
        'weight': opponent_pokemon_1['weight'], 
        } 
    opponent_stats_2 = { 
        'name': opponent_pokemon_2['name'], 
        'id': opponent_pokemon_2['id'], 
        'height': opponent_pokemon_2['height'], 
        'weight': opponent_pokemon_2['weight'], 
        } 

    
    return [opponent_stats_1, opponent_stats_2]


pokemon_opponent_selection()
```
