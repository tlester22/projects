```python
import pandas as pd 
```

```python
#importing the csv into a dataframe, vivendo_claims, making note of existing index
vivendo_claims = pd.read_csv('food_claims_2212.csv', index_col=0)

#checking the output
print(vivendo_claims.head(20))
print(vivendo_claims.info())

#note: missing values in amount_paid and linked cases
#note: claim_amount not numeric type
#note: linked_cases not boolean type
```

```python
                    # time_to_close - no null values,

negative_vals_time = 0

for index, row in vivendo_claims.iterrows():
    if row["time_to_close"] < 0:
        negative_vals_time += 1
        
print('time_to_close negative vaules : ' + str(negative_vals_time))


                    # claim_amount- no null values, 'R$' removed from all rows, column type changed to float


vivendo_claims['claim_amount']=vivendo_claims['claim_amount'].apply(lambda x: x.strip('R$'))
vivendo_claims['claim_amount']=vivendo_claims['claim_amount'].astype(float)

for index, row in vivendo_claims.iterrows():
    if row["claim_amount"] < 0:
        negative_vals_claim_amount += 1

print('claim_amount negative values: ' + str(negative_vals_claim_amount))




                    #amount_paid - 36 null values replaced with median value.

null_vals_amount_paid = 0
negative_vals_amount_paid = 0
median = vivendo_claims['amount_paid'].median()

for index, row in vivendo_claims.iterrows():
    if pd.isnull(row["amount_paid"]) == True:
        null_vals_amount_paid += 1
    if row["amount_paid"] < 0:
        negative_vals_amount_paid += 1
        
print('amount paid null values: ' + str(null_vals_amount_paid))
print('value to replace null values: ' + str(median))
print('amount paid negative values: ' + str(negative_vals_amount_paid))

        
vivendo_claims['amount_paid'] = vivendo_claims['amount_paid'].fillna(value=median, inplace=False)



                # location - no null values, extra values or nonuniform formats detected
null_vals_location = 0
nonuniform_values_location = 0

for index, row in vivendo_claims.iterrows():
    if row['location'] not in ['RECIFE','NATAL','SAO LUIS','FORTALEZA']:
        nonuniform_values_location += 1
        print(row['location'])

print('location values not formatted properly: ' +str(nonuniform_values_location))



               # individuals_on_claim - no null values or other concerns. All values between 1 and 15
negative_vals_individuals = 0

for index, row in vivendo_claims.iterrows():
    if row["amount_paid"] < 0:
        negative_vals_individuals += 1
        
print(negative_vals_individuals)

print('largest number of individuals on a claim: '+ str(max(vivendo_claims['individuals_on_claim'])))
print('smallest number of individuals on a claim: '+ str(min(vivendo_claims['individuals_on_claim'])))



                # linked_cases - 26 null values which were replaced with False, column type changed to bool
null_vals_linked_cases = 0

for index, row in vivendo_claims.iterrows():
    if pd.isnull(row["linked_cases"]) == True:
        null_vals_linked_cases += 1
        
print('linked cases null values: '+ str(null_vals_linked_cases))
print('amount to replace nulls: False')

vivendo_claims['linked_cases'] = vivendo_claims['linked_cases'].fillna(value=False, inplace=False)
vivendo_claims['linked_cases']= vivendo_claims['linked_cases'].astype(bool)




                # cause - 0 null values, 30 values formatted incorrectly, updated to lowercase
wrong_format_cause = 0

for index, row in vivendo_claims.iterrows():
    if row['cause'] not in ['unknown','vegetable','meat']:
        wrong_format_cause += 1

print('cause values not formatted properly: ' +str(wrong_format_cause))

vivendo_claims['cause'] = vivendo_claims['cause'].apply(lambda x: x.lower())
```
